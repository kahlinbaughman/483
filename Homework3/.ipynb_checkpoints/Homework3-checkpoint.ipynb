# Imports

```python
import numpy as np
from scipy import misc
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
from PIL import Image
import math
import cv2
from skimage import filters
from skimage.transform import resize
```

# Question 1

#### Helper Functions

```python

# Converts an image object into an array
def convert(image):
    
    img_arr = np.asarray(image)
    
    return img_arr


# Input a 3d array of iamges returns a single 2d array with the images in a pyramid
def make_pyramid(array):
    
    rows,cols = array[0].shape
    rows_scaled, cols_scaled = array[1].shape
    
    composite = np.zeros((rows, cols+cols_scaled))
    
    composite[:rows, :cols] = array[0]
    
    
    row_tracker = 0
    
    for i in range(1,len(array)):
        composite[row_tracker:row_tracker + (array[i].shape[0]),cols:cols + (array[i].shape[1])] = array[i]
        row_tracker += array[i].shape[0]
    
    return composite
```

## A

```python
# image must be a ndarray
def my_gaussian_pyramid(image, scale_factor=2, n_scales=5, sigma=None):
    
    y,x = image.shape
    pixels = x*y
    max_scales = 0
    
    while pixels != 1:
        pixels = math.ceil(pixels/scale_factor)
        max_scales += 1
        
    n_scales = min(max_scales, n_scales)
    
    images_arr = [0]
    images_arr[0] = image
    
    if(sigma == None):
        sigma = 2 * (scale_factor/6.0)
        
    
    for i in range(1,n_scales):
        temp_arr = gaussian_filter(images_arr[i-1], sigma)
        temp_arr = temp_arr[::scale_factor, ::scale_factor]
        images_arr.append(temp_arr)

    
    return images_arr
```

## B

### 1

```python
butterfly = Image.open('butterfly.png')
butterfly = convert(butterfly)

gaussian_butterfly_sf2 = my_gaussian_pyramid(butterfly, 2, 5)
gaussian_butterfly_sf2 = make_pyramid(gaussian_butterfly_sf2)

plt.imsave("butterfly_pyramid_sf2.jpg", gaussian_butterfly_sf2, cmap='gray')
plt.imshow(gaussian_butterfly_sf2, cmap='gray')
```

### 2

```python
gaussian_butterfly_sf3 = my_gaussian_pyramid(butterfly, 3, 10, 5)
gaussian_butterfly_sf3 = make_pyramid(gaussian_butterfly_sf3)

plt.imsave("butterfly_pyramid_sf3_sigma5.jpg", gaussian_butterfly_sf3, cmap='gray')
plt.imshow(gaussian_butterfly_sf3, cmap='gray')
```

## C

```python
gaussian_butterfly_edges = my_gaussian_pyramid(butterfly, 2, 5)

for i in range(len(gaussian_butterfly_edges)):
    gaussian_butterfly_edges[i] = filters.sobel(gaussian_butterfly_edges[i])

gaussian_butterfly_edges = make_pyramid(gaussian_butterfly_edges)

plt.imsave("butterfly_pyramid_edges.jpg", gaussian_butterfly_edges, cmap='gray')
plt.imshow(gaussian_butterfly_edges, cmap='gray')
```

## D

```python
gaussian_butterfly_0 = my_gaussian_pyramid(butterfly)
edges_butterfly_0 = [0]
edges_butterfly_0[0] = filters.sobel(gaussian_butterfly_0[0])

for i in range(1, len(gaussian_butterfly_0)):
    temp = filters.sobel(gaussian_butterfly_0[i])
    edges_butterfly_0.append(temp)
    
for i in range(1, len(gaussian_butterfly_0)):
    gaussian_butterfly_0[i] = resize(gaussian_butterfly_0[i], gaussian_butterfly_0[0].shape)
    
for i in range(1, len(gaussian_butterfly_0)):
    edges_butterfly_0[i] = resize(edges_butterfly_0[i], edges_butterfly_0[0].shape)
    
```

```python
edge0 = plt.figure()
edge0.set_figheight(20)
edge0.set_figwidth(20)

axes0_1 = edge0.add_subplot(131)
axes0_1.get_xaxis().set_visible(False)
axes0_1.get_yaxis().set_visible(False)
axes0_1.set_title("Input Image")
axes0_1.imshow(gaussian_butterfly_0[0], cmap='gray')

axes0_2 = edge0.add_subplot(132)
axes0_2.get_xaxis().set_visible(False)
axes0_2.get_yaxis().set_visible(False)
axes0_2.set_title("Sobel edge detection at level 0")
axes0_2.imshow(edges_butterfly_0[0], cmap='gray')

edge0.savefig("butterfly_pyramid_edge0.jpg")
```
